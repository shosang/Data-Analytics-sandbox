```python
import pandas as pd
```

```python
import matplotlib as mpl

mpl.rcParams['text.color'] = 'w'
mpl.rcParams['xtick.color'] = 'w'
mpl.rcParams['ytick.color'] = 'w'
mpl.rcParams['axes.labelcolor'] = 'w'
```

```python
# Read in data
data_red = pd.read_csv("winequality-red.csv", sep=";")
data_red.head(5)
```

```python
# Summary statistics for numerical variables
data_red.describe()
```

```python
# Create histograms for all numeric variables
data_red.hist(figsize=(10,10))
plt.show()
```

```python
# Quality can be considered a categorical (ordinal) variable
data_red['quality'].value_counts()
```

```python
# Create a frequency graph for the quality variable
data_red['quality'].value_counts().plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```
