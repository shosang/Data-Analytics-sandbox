```python
# Find all NA values in dataset
print(df.isna().sum())
```

Missing data does not always show up as NaN/NULL values. <br>
Sometimes a specific value is chosen to represent missing data in your database.<br>
An understanding of what your data represents is important here. 

For the diabetes dataset, the value of 0 does not make sense for these columns:<br>
1. Glucose
2. BloodPressure
3. SkinThickness
4. Insulin
5. BMI


```python
print('# values missing in Glucose column: ' + str(len(df[df.Glucose == 0])))
print('# values missing in BloodPressure column: ' + str(len(df[df.BloodPressure == 0])))
print('# values missing in SkinThickness column: ' + str(len(df[df.SkinThickness == 0])))
print('# values missing in Insulin column: ' + str(len(df[df.Insulin == 0])))
print('# values missing in BMI column: ' + str(len(df[df.BMI == 0])))
```

```python
# Replace values = 0 from these 5 columns with NA
df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].replace(0, np.NaN)
df.head(5)
```

```python
df.describe()
```

```python
# Drop insulin column - too many missing values
df = df.drop('Insulin', axis=1)
df.head(5)
```

```python
# BloodPressure, BMI, Glucose - remove rows that are missing
df = df[(df.BloodPressure != 0) &
        (df.BMI != 0) &
        (df.Glucose != 0)]
df.head(5)
```

```python
import matplotlib.pyplot as plt
df.hist('SkinThickness')
plt.axvline(df.SkinThickness.mean(), color='r')
plt.axvline(df.SkinThickness.median(), color='g')
```

```python
# Replace missing values from SkinThickness column with mean value
df.SkinThickness.fillna(df.SkinThickness.mean(), inplace=True)
df.head(5)
```
